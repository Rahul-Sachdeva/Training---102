# Pandas Worksheet

**Import pandas**

```python
import pandas as pd
import numpy as np
```

**Read the Ecommerce Purchases csv file and store it in a dataframe**

```python
df = pd.read_csv('/content/Ecommerce Purchases.csv')
df
```

**Check the head of the dataframe**

```python
df.head()
```

**How many rows and columns are there in the dataframe?**

```python
df.shape
```

**What is the average Purchase Price?**

```python
df['Purchase Price'].mean()
```

**What were the highest and lowest purchase prices?**

```python
df['Purchase Price'].max()
```

```python
df['Purchase Price'].min()
```

**How many people have English 'en' as their language of choice on the website?**

```python
df[df['Language'] == 'en']["Language"].count()
```

**How many people have the job title of Lawyer?**

```python
df['Job'].value_counts()['Lawyer']
```

**How many people made the purchases during AM and PM?**

```python
df['AM or PM'].value_counts()['AM' and 'PM']
```

**What are the five most common job titles?**

```python
df['Job'].value_counts().head()
```

**Someone made a purchase that came from Lot: "90 WT". What was the Purchase Price for this transaction?**

```python
df[df['Lot'] == '90 WT']['Purchase Price']
```

**How many people have American Express as their Credit Card Provider *and* made a purchase above $95?**

```python
df[(df['CC Provider']=='American Express') & (df['Purchase Price']>95)]['Purchase Price'].count()
```

**Hard: How many people have a credit card that expires in 2025?**

```python
df[df['CC Exp Date'].str.contains('/25' )]['CC Exp Date'].count()
```

**Hard: What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...)**

```python
df['Email'].apply(lambda x: x.split('@')[1]).value_counts().head()
```
